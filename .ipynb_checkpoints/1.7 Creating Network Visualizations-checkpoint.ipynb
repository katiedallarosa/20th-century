### 1.7: Creating Network Visualizations

In this task, you’ll use your data on the relationships between countries that played a key role in twentieth-century events to build and analyze a network. You’ll also look into the groups of allies in the history of the 20th century.

```python
# Import libraries

import pandas as pd
import numpy as np
import networkx as nx
import os
import matplotlib.pyplot as plt
from pyvis.network import Network
import seaborn as sns
import pyvis
```

#### Q2. Import the country relationships file as a dataframe

```python
# Import relationship data

relationship_df = pd.read_csv('countries_relationship.csv', index_col = 0)
```

```python
relationship_df
```

#### Q3. Create a network object from the countries relationships dataframe.

```python
# Create a graph from a pandas dataframe

G = nx.from_pandas_edgelist(relationship_df, 
                            source = "source", 
                            target = "target", 
                            edge_attr = "value", 
                            create_using = nx.Graph())
```

#### Q4. Create a static network graph using only NetworkX.

```python
h = plt.figure(figsize = (7,7))
pos = nx.kamada_kawai_layout(G)
nx.draw(G, with_labels = True, node_color = 'red', edge_cmap = plt.cm.Blues, pos = pos)
plt.show() 
```

```python
h.savefig('networkx_plt_countries.png')
```

#### Q5. Switch to pyvis, build an interactive network graph from your network object, and save it as an HTML file

```python
# Define net object

net = Network()
net = Network(notebook = True,width="900px", height="800px", bgcolor='#222222', font_color='white')

# Define nodes from the G object
node_degree = dict(G.degree)

# Setting up node size attributes
nx.set_node_attributes(G, node_degree, 'size')
net.from_nx(G)
net.show_buttons(filter_ = True)
net.repulsion()
net.show("20thcentury.html")
```

#### Q6. Apply the Leiden algorithm to detect any communities within the countries.

```python
from cdlib import algorithms
coms = algorithms.leiden(G)
```

```python
type(coms)
```

```python
new_com = coms.to_node_community_map()
```

```python
new_com
```

#### Q7. Visualize the output of the communities algorithm on a new network graph and save it as HTML. Write your observations in a Markdown cell. Does the partition make sense according to history? What other conclusions can you make about the connections between the communities?

```python
# Put the output from new_com into a dictionary

dict_com = {k:v[0] for k,v in new_com.items()}
```

```python
dict_com
```

```python
# Rerun network graph with the communities as an argument

nx.set_node_attributes(G, dict_com, 'group')
com_net = Network(notebook = True, width="1300px", height="700px", bgcolor='#222222', font_color='white')
com_net.from_nx(G)
com_net.show_buttons(filter_ = True)
com_net.repulsion()
com_net.show("countries_communities_leiden.html")
```

#### Discussion

It's not surprising to see that Germany is shown as a central factor for the 20th Century key events and within this graph. We know many events tied into Germany such as WWI, WWII, The Holocaust, The Cold War and the fall of the Berlin Wall. Other key countries include Japane, France, and Italy. 

#### Q8. Calculate and visualize the degree, closeness, and betweenness centrality measures covered in this Exercise. In the markdown cells underneath, note your observations about the outputs.

```python
# Degree centrality

degree_dict = nx.degree_centrality(G)
degree_dict
```

```python
degree_df = pd.DataFrame(degree_dict.items(), columns=['countries','centrality'])
```

```python
degree_df.sort_values(by = ['centrality'], ascending=False, inplace = True)
```

```python
# Plot the degree centrality

plt.figure(figsize = (15, 10))
with sns.dark_palette("xkcd:blue", 22):
    sns.barplot(x = "centrality", y = "countries",
    saturation = 0.9, data = degree_df).set_title("20th Century Countries - degree centrality")
```

```python
# Closeness centrality

closeness_dict = nx.closeness_centrality(G)
closeness_df = pd.DataFrame(closeness_dict.items(), columns=['countries','centrality'])
```

```python
closeness_df.sort_values(by = ['centrality'], ascending=False, inplace = True)
```

```python
plt.figure(figsize = (15, 10))
with sns.dark_palette("xkcd:blue", 22):
    sns.barplot(x = "centrality", y = "countries",
    saturation = 0.9, data = closeness_df).set_title("20th Century Countries - closeness centrality")
```

```python
# Betweenness centrality

betweenness_dict = nx.betweenness_centrality(G)
betweennes_df = pd.DataFrame(betweenness_dict.items(), columns=['countries','centrality'])
```

```python
betweennes_df.sort_values(by = ['centrality'], ascending=False, inplace = True)
```

```python
plt.figure(figsize = (15, 10))
with sns.dark_palette("xkcd:blue", 22):
    sns.barplot(x = "centrality", y = "countries",
    saturation = 0.9, data = betweennes_df).set_title("20th Century Countries - betweenness centrality")
```

#### Discussion

Degree centrality: Germany has the highest degree centrality, which shows us they had the most direct connections with other countries. Countries like Austria and Spain have lower values, which would mean fewer connections.

Closeness centrality: We can see that Germany and Japan had higher closeness centrality and are strategically located within the graph. We can assume this allowed them to influence or access other countries more efficiently.

Betweeness centrality: Again, Germany has a high betweeness centrality which shows they played a crucial role between the other countries in the 20th century and it's key eventns.

```python
# Save centrality measures

nx.set_node_attributes(G, degree_dict, 'degree_centrality')
nx.set_node_attributes(G, betweenness_dict, 'betweenness_centrality')
nx.set_node_attributes(G, closeness_dict, 'closeness_centrality')
```
